```python
import pyttsx3
import speech_recognition as sr
import datetime
import webbrowser
import os



engine = pyttsx3.init()
engine.setProperty("rate", 170)      # Default speaking speed
engine.setProperty("volume", 1.0)    # Max volume

def speak(text):
    print("Assistant:", text)
    engine.say(text)
    engine.runAndWait()


listener = sr.Recognizer()

def listen():
    """Listens to user's voice and converts into text."""
    try:
        with sr.Microphone() as source:
            listener.adjust_for_ambient_noise(source)
            print("\nListening...")
            audio = listener.listen(source)
            command = listener.recognize_google(audio, language="en-IN")
            command = command.lower()
            print("You:", command)
            return command
    except:
        return ""



def greet_user():
    hour = int(datetime.datetime.now().hour)
    if 5 <= hour < 12:
        speak("Good morning Yeeshu")
    elif 12 <= hour < 17:
        speak("Good afternoon Yeeshu")
    else:
        speak("Good evening Yeeshu")

    speak("How can I help you today?")

    

def run_assistant():
    greet_user()

    while True:
        command = listen()

   
        if "stop" in command or "exit" in command or "bye" in command:
            speak("Okay, shutting down. Take care.")
            break

       
        elif "slow" in command:
            engine.setProperty("rate", 140)
            speak("Done, I will speak a little slower now.")

        elif "fast" in command:
            engine.setProperty("rate", 200)
            speak("Now I will speak a bit faster.")

        # Website opening
        elif "youtube" in command:
            speak("Opening YouTube.")
            webbrowser.open("https://youtube.com")

        elif "google" in command:
            speak("Opening Google.")
            webbrowser.open("https://google.com")

        # Time
        elif "time" in command:
            now = datetime.datetime.now().strftime("%I:%M %p")
            speak(f"The time is {now}")

      
        elif "save this" in command:
            with open("saved_text.txt", "a") as f:
                f.write(command + "\n")
            speak("I saved what you said.")

     
        elif "read file" in command:
            try:
                with open("saved_text.txt", "r") as f:
                    content = f.read()
                speak("Here is what is written in the file.")
                speak(content)
            except:
                speak("Looks like the file is empty or missing.")

   
        elif "hindi me bolo" in command:
            speak("ठीक है, अब मैं हिंदी में बोलूँगा।")
            engine.setProperty("voice", "hi")

        elif "english me bolo" in command:
            engine.setProperty("voice", "")
            speak("Alright, I am back to English.")

        # Fall-back
        else:
            speak("I heard you, but I am not sure what to do with that.")

# Run the assistant
run_assistant()
```
