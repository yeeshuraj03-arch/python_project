```python
print("===== Welcome to Movie Ticket Booking System =====\n")

movies = {
    1: {"name": "Avengers: Endgame", "price": 250},
    2: {"name": "KGF Chapter 2", "price": 200},
    3: {"name": "Pushpa 2", "price": 180},
    4: {"name": "Jawan", "price": 220},
}

while True:
    # Age check
    age = int(input("Enter your age: "))

    if age < 18:
        print("You must be 18 or above to watch these movies. Booking not allowed.")
        continue

    print("\nAvailable Movies:")
    for key, value in movies.items():
        print(f"{key}. {value['name']} - ₹{value['price']} per ticket")

    choice = int(input("\nSelect a movie (1-4): "))

    if choice in movies:
        movie = movies[choice]
        print(f"\nYou selected: {movie['name']}")

        seats = int(input("Enter number of tickets: "))
        total = seats * movie["price"]

        print("\n===== Booking Summary =====")
        print(f"Movie: {movie['name']}")
        print(f"Tickets: {seats}")
        print(f"Price per ticket: ₹{movie['price']}")
        print(f"Total Amount: ₹{total}")
        print("============================")

        confirm = input("Confirm booking? (yes/no): ").lower()

        if confirm == "yes":
            print("Tickets booked successfully. Enjoy the movie.")
        else:
            print("Booking cancelled.")

    else:
        print("Invalid choice. Please select a movie from 1 to 4.")

    print("\nThank you for using Movie Ticket Booking System. Goodbye!")
    break
```
