```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import time
from datetime import datetime

def login():
    st.title("üîê Login Page")

    username = st.text_input("Enter username")
    password = st.text_input("Enter password", type="password")

    if st.button("Login"):
        if username == "admin" and password == "123":
            st.success("Login Successful!")
            st.session_state["logged_in"] = True
            time.sleep(1)
            st.rerun()
        else:
            st.error("Invalid username or password")
            

def home_page():
    st.title("üè† Home Page")
    
    
    st.subheader("‚è∞ Current Time & Date")
    st.info(datetime.now().strftime("%Y-%m-%d  |  %H:%M:%S"))

    
    mode = st.selectbox("Choose Theme:", ["Light Mode", "Dark Mode"])
    if mode == "Dark Mode":
        st.markdown("<style>body{background-color:#111;color:white;}</style>", unsafe_allow_html=True)

    
    st.subheader("üìä Loading Progress")
    progress = st.progress(0)
    for i in range(100):
        progress.progress(i + 1)
        time.sleep(0.01)

    st.success("Loaded Successfully!")


def student_data_page():
    st.title("üìò Student Data + Search Filter")

    data = {
        "Name": ["Raju", "Aman", "Neha", "Priya", "Rohan"],
        "Marks": [85, 92, 76, 88, 90]
    }

    df = pd.DataFrame(data)

    search = st.text_input("Search student name")

    if search:
        filtered = df[df["Name"].str.contains(search, case=False)]
        st.write(filtered)
    else:
        st.write(df)


# PAGE 3 - CHARTS PAGE

def chart_page():
    st.title("üìä Charts & Visualizations")

    data = {
        "Name": ["Raju", "Aman", "Priya", "Neha"],
        "Marks": [85, 92, 78, 88]
    }

    df = pd.DataFrame(data)

    st.subheader("Bar Chart")
    fig, ax = plt.subplots()
    ax.bar(df["Name"], df["Marks"])
    st.pyplot(fig)

    st.subheader("Line Chart")
    fig2, ax2 = plt.subplots()
    ax2.plot(df["Name"], df["Marks"])
    st.pyplot(fig2)

    st.subheader("Pie Chart")
    fig3, ax3 = plt.subplots()
    ax3.pie(df["Marks"], labels=df["Name"], autopct="%1.1f%%")
    st.pyplot(fig3)


def main_app():
    st.sidebar.title("üìå MENU")

    page = st.sidebar.radio("Select Page", ["Home", "Student Data", "Charts", "Logout"])

    if page == "Home":
        home_page()
    elif page == "Student Data":
        student_data_page()
    elif page == "Charts":
        chart_page()
    elif page == "Logout":
        st.session_state["logged_in"] = False
        st.rerun()


if "logged_in" not in st.session_state:
    st.session_state["logged_in"] = False

if st.session_state["logged_in"]:
    main_app()
else:
    login()
```
